```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/sample_data/cars.csv')
```

```python
print(df.head())
print(df.dtypes)
```

```python
df['hp'].fillna(df['hp'].mean(), inplace=True)
df.dropna(subset=['hp'], inplace=True) # Eliminar filas.
```

```python
df_filtered = df[df['cyl'] == 6]
print(df_filtered)
```

```python
def categorize_hp(hp):
if hp > 150:
return 'High'
elif hp > 80:
return 'Medium'
else:
return 'Low'
df_filtered['HP_Category'] = df_filtered['hp'].apply(categorize_hp)
```

```python
df_filtered.to_csv('.\dataset\cleaned_cars_dataset.csv', index=False)
```
